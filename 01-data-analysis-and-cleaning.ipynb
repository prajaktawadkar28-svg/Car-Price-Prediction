```python
#Lets clean year column
#change datatype - put null at non-changable value
df['year'] = pd.to_numeric(df['year'], errors = 'coerce')
df.dropna(subset = ['year'], inplace = True)
df['year'] = df['year'].astype(int)
```

```python
#Lets clean Price column
# for i in df.index:
#     df.loc[i, 'Price'] = df.loc[i, 'Price'].replace(',', '')
df['Price']  = df['Price'].str.replace(',', '')
df['Price'] = pd.to_numeric(df['Price'], errors = 'coerce')
df.dropna(subset = ['Price'], inplace = True)
df['Price'] = df['Price'].astype(int)
```

```python
df['kms_driven']  = df['kms_driven'].str.replace(',', '')
df['kms_driven']  = df['kms_driven'].str.replace(' kms', '')
df['kms_driven'] = pd.to_numeric(df['kms_driven'], errors = 'coerce')
df.dropna(subset = ['kms_driven'], inplace = True)
df['kms_driven'] = df['kms_driven'].astype(int)
```

```python
df.dropna(inplace = True)
```

```python
df['company'].unique()
```

```python
#remove company name from name and take only 2 words
for i in df.index:
    name = df.loc[i, 'name']
    name = ' '.join(name.split(' ')[1:3])
    df.loc[i, 'name'] = name
```
